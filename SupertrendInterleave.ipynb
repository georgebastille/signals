```python
import pickle
import pandas as pd
from indicators.indicator.indicators import SuperTrend
```

```python
df = pickle.load(open('fxcm_eur_usd_12_hr_1m', 'rb'))
ohlc_rename = {'bidopen':'Open', 'bidhigh':'High', 'bidlow':'Low', 'bidclose': 'Close'}
df = df.rename(columns=ohlc_rename)
ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
df = df.resample(f"30Min", base=0).apply(ohlc_dict).dropna()

#display(df)
```

```python
def meta_super_trend(df, sample_mins, period, multiplier):

    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    base_interval = (df.index[1]-df.index[0]).components.minutes
    dfs = []
    for base in range(0, sample_mins, base_interval):
        tmp = df.resample(f"{sample_mins}Min", base=base).apply(ohlc_dict).dropna()
        SuperTrend(tmp, period, multiplier, prefix=sample_mins)
        dfs.append(tmp)

    return pd.concat(dfs).sort_index()
```

```python
df60 = meta_super_trend(df, sample_mins=240, period=3, multiplier=2)
display(df60)
```

```python
# How to combine multiple supertrends into one dataframe
df60 = meta_super_trend(df, 60, 3, 2)
df30 = meta_super_trend(df, 30, 3, 2)
cols_to_use = df60.columns.difference(df30.columns)
dfNew = pd.merge(df30, df60[cols_to_use], left_index=True, right_index=True, how='inner')


display(dfNew)
```
